# pandas GroupBy function:

```python
# what is pandas GroupBy Function?
# pandas groupby function is used to split the data into groups based on some criteria:
# Syntax: DataFrame.groupby()


# Any groupby operation involves one of the following operation on the original object;
# - Splitting the object
# - Applying a function
# - Combining the result
```

```python
import pandas as pd
```

```python
df = pd.read_csv("E:\\New folder (4)\\Book1.csv")
df
```

```python
gr1=df.groupby(by = "Date")
gr1
```

```python
gr1.groups
```

```python
gr1=df.groupby(["Date","sleeping h"]).groups
gr1
```

```python
for key,value in gr1:
    print(key)
    print(value)
```

```python
gr1=df.groupby(by = "Date")
gr1
```
