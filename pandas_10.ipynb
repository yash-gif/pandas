```python
import pandas as pd
```

```python
df = pd.read_csv("E:\\New folder (4)\\Book2.csv")
df
```

```python
# drop_na : drop the column which contain null values or NaN values and create a new dataframe.
```

```python
df.dropna()
```

```python
df.dropna(axis = 1)
```

```python
df.dropna(how = "any")# default
```

```python
df.dropna(how = "any",axis = 1)
```

```python
df.dropna(how = "all")# if all values in the adatframe are null then it drops all columns and rows .
```

```python
df.dropna(how = "all" , axis = 1) # it drops column because all values in the column are null values.
```

```python
df.dropna(thresh = 1)# print those rows which contain  one non - null value.
```

```python
df.dropna(thresh = 1 , axis = 1) # print those columns which contain one non - null value
```
