```python
import pandas as pd
```

```python
# what is pandas fillna() method?
# fillna fills the NaN values with given values (input)
# syntax : DataFrame.fillna()
```

```python
df = pd.read_csv("E:\\New folder (4)\\Book2.csv")
df
```

```python
df.fillna(0)
```

```python
df.fillna(2)
```

```python
df.fillna({"Name":"None","Companies":"None","Growth":0})
```

```python
df.fillna(method = "ffill")#fills NaN values with forward value:
```

```python
df.fillna(method = "pad")# do same work as above function does:
```

```python
df.fillna(method = "bfill")# fills NaN values with backward value:
```

```python
df.fillna(axis = 0 , method = "ffill")
```

```python
df.fillna(axis = 1 , method = "ffill") # fills previous column values in place of NaN
```

```python
df.fillna(axis = 1 , method = "bfill")
```

```python
df.fillna(0 , limit = 5)
```

```python
df.fillna(method = "ffill" , limit = 5 , ) # apply by default on rows and limit should be greater then 0:
```

```python
df.fillna(method = "ffill" , limit = 5 , axis = 1 ) # by using axis = 1 it applies on columns:
```

```python
df.fillna(6,inplace = True ) # it update the current dataframe:
df
```
