```python
# what is pandas concat() function?
# pandas provides various facilities for easily combining together Series,DataFrame and Panel objects.
# syntax: Pandas.concat()
# parameters : objs , ignore_index , axis , keys , join , sort ,join_axes
```

```python
import pandas as pd
```

```python
sr1 =pd.Series([0,1,2])
sr1
```

```python
sr2 =pd.Series([3,4,5,6,7,8])
sr2
```

```python
pd.concat([sr1 , sr2]) # Series , DataFrame or Panel objects whatever you use for input you have to use list.
```

```python
df1 =pd.DataFrame({"ID":[1,2,3,4] , "Name":["A","B","C","D"],"Class":[5,6,7,8]})
df1
```

```python
df2 =pd.DataFrame({"ID":[5,6,7,8] , "Name":["E","F","G","H"],"Class":[9,10,11,12]})
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df2,df1]) # axis parameter by default works as axis = 0(index or row)
```

```python
pd.concat([df1,df2],axis =1)
```

```python
pd.concat([df1,df2],axis =0,ignore_index=True) # for correct systametic index.
```

```python
df1 =pd.DataFrame({"ID":[1,2,3,4] , "Name":["A","B","C","D"],"Class":[5,6,7,8]})
df1
```

```python
df2 =pd.DataFrame({"ID":[3,4] , "Name":["C","D"],"Class":[7,8]})
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2] , axis = 1)
```

```python
pd.concat([df1,df2] , axis = 1,join = "inner") # print intersection value of both DataFrame.
```

```python
pd.concat([df1,df2] , axis = 1,join_axes = [df2.index]) 
```
